#### Nun mache ich einen Loop über alle Stunden, erstelle einen DF mit Percentilen pro Stunde, und bilde aus diesen 24 DFs eine "List of Dictionaries" für Flow und Speed. Auf Basis dieser Dictionaries kann ich schliesslich DFs in dem von mir gewünschten Format konstruieren.

```python
ListPercentileFlow = []
ListPercentileSpeed = []
DfPercentileFlow = pd.DataFrame()
DfPercentileSpeed = pd.DataFrame()
for h in range(24):
    q = db.select(t.c.time, func.strftime('%Y', t.c.time).label('Jahr'), func.strftime('%m', t.c.time).label('Monat'), \
              func.strftime('%W', t.c.time).label('Woche'), func.strftime('%d', t.c.time).label('Tag'), func.strftime('%H', t.c.time, 'localtime').label('Stunde'), \
               func.strftime('%w', t.c.time).label('Wochentag'), func.round((t.c.vfr11)).label('flow'), func.round((t.c.s12)).label('speed')). \
    filter(and_(t.c.vfr11 > 0, cast(func.strftime('%H', t.c.time, 'localtime'), Integer) == h))
    result = conn.execute(q)
    row = result.fetchall()
    df_help= pd.read_sql_query(q, conn)
    perc = df_help.quantile([0.05, .1, .3, .5, .7, .9, .95], axis = 0, interpolation = "higher")
    flow = {"hour" : h}
    flow.update(perc["flow"].to_dict())
    ListPercentileFlow.append(flow)
    speed = {"hour" : h}
    speed.update(perc["speed"].to_dict())    
    ListPercentileSpeed.append(speed)
    
```

```python
df_perc_flow = pd.DataFrame.from_dict(ListPercentileFlow)
df_perc_flow.to_csv(r'C:\Users\peter\Coding\Jupyter\Grafiken\Test_data_flow' + '.csv', index = False)
df_perc_flow
```
